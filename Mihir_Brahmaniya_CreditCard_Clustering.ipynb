# Load & Inspect the Credit Card Dataset

```python
# Task 1 â€” Load and Inspect Dataset
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("/content/CC GENERAL.csv")
print("âœ… Data Loaded Successfully!")
print("Shape:", data.shape)
print("Columns:", data.columns.tolist())

# Preview
display(data.head())

# Summary statistics
display(data.describe())

# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())

# Drop CUST_ID (non-numeric identifier)
if 'CUST_ID' in data.columns:
    data.drop('CUST_ID', axis=1, inplace=True)
    print("âœ… Dropped 'CUST_ID' column.")

# Fill missing values with column means
data.fillna(data.mean(), inplace=True)
print("\nâœ… Missing values handled by replacing with mean.")


# Save cleaned dataset
data.to_csv("CC_GENERAL_cleaned.csv", index=False)
print("\nâœ… Cleaned data saved as 'CC_GENERAL_cleaned.csv'")
```

# Data Preprocessing and Standardization

```python
from sklearn.preprocessing import StandardScaler

# Load cleaned data
df = pd.read_csv("CC_GENERAL_cleaned.csv")

# Standardize numerical columns
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Convert back to DataFrame for readability
X_scaled_df = pd.DataFrame(X_scaled, columns=df.columns)

# Save standardized dataset
X_scaled_df.to_csv("CC_GENERAL_scaled.csv", index=False)
print("âœ… Data standardized and saved as 'CC_GENERAL_scaled.csv'")

# Quick check
X_scaled_df.describe().T.head()
```

# Determining the Optimal Number of Clusters

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X_scaled = pd.read_csv("CC_GENERAL_scaled.csv")

wcss, sil_scores, bic_scores = [], [], []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X_scaled)
    wcss.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, kmeans.labels_))

    gmm = GaussianMixture(n_components=k, random_state=42).fit(X_scaled)
    bic_scores.append(gmm.bic(X_scaled))

# Plot results
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(K, wcss, 'bo-')
plt.title('Elbow Method (WCSS)')
plt.xlabel('k'); plt.ylabel('WCSS')

plt.subplot(1,3,2)
plt.plot(K, sil_scores, 'ro-')
plt.title('Silhouette Score')
plt.xlabel('k'); plt.ylabel('Score')

plt.subplot(1,3,3)
plt.plot(K, bic_scores, 'go-')
plt.title('BIC (Gaussian Mixture Model)')
plt.xlabel('k'); plt.ylabel('BIC')

plt.tight_layout()
plt.show()
```

# AIC, AICc, and BIC for Optimal Number of Clusters

```python
n, d = X_scaled.shape
aic_vals, aicc_vals, bic_vals = [], [], []
K = range(2, 11)

for k in K:
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
    gmm.fit(X_scaled)

    aic = gmm.aic(X_scaled)
    bic = gmm.bic(X_scaled)

    num_params = (k - 1) + (k * d) + (k * d * (d + 1) / 2)
    aicc = aic + (2 * num_params * (num_params + 1)) / (n - num_params - 1)

    aic_vals.append(aic)
    aicc_vals.append(aicc)
    bic_vals.append(bic)

plt.figure(figsize=(8,5))
plt.plot(K, aic_vals, 'r--', label='AIC')
plt.plot(K, aicc_vals, 'b-', label='AICc (Corrected)')
plt.plot(K, bic_vals, 'g-', label='BIC')
plt.title('Model Selection Criteria')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Criterion Value')
plt.legend()
plt.grid(True)
plt.show()

print(f"Best k by AIC: {K[np.argmin(aic_vals)]}")
print(f"Best k by AICc: {K[np.argmin(aicc_vals)]}")
print(f"Best k by BIC: {K[np.argmin(bic_vals)]}")
```

# Hierarchical Clustering and Dendrogram Cutting Methods

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.metrics import silhouette_score
import numpy as np

link_matrix = linkage(X_scaled, method='ward')

# Dendrogram
plt.figure(figsize=(10,6))
dendrogram(link_matrix, truncate_mode='lastp', p=20, leaf_rotation=45., leaf_font_size=10.)
plt.title('Hierarchical Clustering Dendrogram (Ward)')
plt.xlabel('Sample index'); plt.ylabel('Distance')
plt.show()

# Distance jump method
distances = link_matrix[:, 2]
max_jump_idx = np.argmax(np.diff(distances))
cut_distance = distances[max_jump_idx]
labels_dist = fcluster(link_matrix, t=cut_distance, criterion='distance')
print(f"Best cut distance (max jump): {cut_distance:.3f}")

# Silhouette-based
sil_scores = []
for k in range(2, 11):
    labels = fcluster(link_matrix, k, criterion='maxclust')
    sil_scores.append(silhouette_score(X_scaled, labels))

best_k_sil = np.argmax(sil_scores) + 2
plt.plot(range(2,11), sil_scores, 'bo-')
plt.title('Silhouette Scores vs Cluster Count')
plt.xlabel('Clusters (k)')
plt.ylabel('Silhouette')
plt.show()

print(f"Best k by silhouette for hierarchical clustering: {best_k_sil}")
```

# Compute Clustering Evaluation Metrics

```python
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score, adjusted_rand_score, normalized_mutual_info_score, v_measure_score

# Choose k (based on previous results)
best_k = 5
kmeans = KMeans(n_clusters=best_k, random_state=42).fit(X_scaled)
df['Cluster'] = kmeans.labels_

# Simulate labels from CREDIT_LIMIT
df['Sim_Label'] = pd.qcut(df['CREDIT_LIMIT'], q=best_k, labels=False)

# Purity
def purity(y_true, y_pred):
    contingency = pd.crosstab(y_true, y_pred)
    return np.sum(np.max(contingency.values, axis=0)) / np.sum(contingency.values)

# Compute metrics
pur = purity(df['Sim_Label'], df['Cluster'])
sil = silhouette_score(X_scaled, df['Cluster'])
db = davies_bouldin_score(X_scaled, df['Cluster'])
ch = calinski_harabasz_score(X_scaled, df['Cluster'])
ari = adjusted_rand_score(df['Sim_Label'], df['Cluster'])
nmi = normalized_mutual_info_score(df['Sim_Label'], df['Cluster'])
vmeas = v_measure_score(df['Sim_Label'], df['Cluster'])

metrics = pd.DataFrame({
    'Metric': ['Purity', 'Silhouette', 'DB Index', 'CH Index', 'ARI', 'NMI', 'V-Measure'],
    'Score': [pur, sil, db, ch, ari, nmi, vmeas]
})

display(metrics)
```

# Anomaly Detection using IOF and LOF

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
import seaborn as sns

# Isolation Forest
iso = IsolationForest(contamination=0.03, random_state=42)
df['IOF_Anomaly'] = np.where(iso.fit_predict(X_scaled) == -1, 1, 0)

# Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.03)
df['LOF_Anomaly'] = np.where(lof.fit_predict(X_scaled) == -1, 1, 0)

# Compare anomalies
iof_count = df['IOF_Anomaly'].sum()
lof_count = df['LOF_Anomaly'].sum()
overlap = ((df['IOF_Anomaly']==1) & (df['LOF_Anomaly']==1)).sum()

summary = pd.DataFrame({
    'Method': ['Isolation Forest', 'Local Outlier Factor', 'Overlap (Both)'],
    'Anomalies Found': [iof_count, lof_count, overlap],
    'Percent of Data': [
        round(iof_count/len(df)*100,2),
        round(lof_count/len(df)*100,2),
        round(overlap/len(df)*100,2)
    ]
})
display(summary)

# Visualize anomalies
plt.figure(figsize=(7,5))
sns.scatterplot(x=df['BALANCE'], y=df['PURCHASES'], hue=df['IOF_Anomaly'], palette=['blue','red'])
plt.title('Isolation Forest (Red = Anomaly)')
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(x=df['BALANCE'], y=df['PURCHASES'], hue=df['LOF_Anomaly'], palette=['green','red'])
plt.title('Local Outlier Factor (Red = Anomaly)')
plt.show()

# Save final dataset
df.to_csv("CC_GENERAL_final_results.csv", index=False)
print("âœ… Final dataset saved as 'CC_GENERAL_final_results.csv'")
```
